```python
f = open("corr","r")
interest = ["Hexim1","Ceacam1","Ppp1r10"]
dataDict = {}
    
for line in f:
    line = line.split(" ")
    if line[0] in interest:
        if line[1] not in dataDict:
            dataDict[line[1]] = {}
        dataDict[line[1]][line[0]] = float(line[2])
dataDict["Hexim1"]["Hexim1"] = 1
dataDict["Ceacam1"]["Ceacam1"] = 1
dataDict["Ppp1r10"]["Ppp1r10"] = 1


print(dataDict)
        
```

```python
import seaborn as sns
#%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.DataFrame.from_dict(dataDict,orient='index',columns = interest)
df.head

```

```python
sns.heatmap(df, cmap='RdYlGn_r', linewidths=0.5, annot=True)
```
